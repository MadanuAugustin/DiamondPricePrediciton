```python
## displaying the number of categorical and numerical features from the dataset

numerical_features = [feature for feature in copydata.columns if copydata[feature].dtypes != 'object']

numerical_features
```

```python
## finding the sub_categories in the categorical features

sub_cat_cut = copydata.cut.unique()
sub_cat_color = copydata.color.unique()
sub_cat_clarity = copydata.clarity.unique()
```

```python
fig, ax = plt.subplots(1, 3, figsize =(20,7))

plt.subplot(131)
plt.title('Value counts of CUT')
plt.pie(copydata.cut.value_counts(), labels =['Ideal', 'Premium', 'Very Good', 'Good', 'Fair'], autopct = '%.2f%%')

plt.subplot(132)
plt.title('Value counts of COLOR')
plt.pie(copydata.color.value_counts(), labels = ['G', 'E', 'F', 'H', 'D', 'I', 'J'], autopct = '%.2f%%')

plt.subplot(133)
plt.title('Value counts of CLARITY')
plt.pie(copydata.clarity.value_counts(), labels=['SI1', 'VS2', 'SI2', 'VS1', 'VVS2', 'VVS1', 'IF', 'I1'], autopct ='%.2f%%')
plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize = (20, 7))

plt.subplot(131)
plt.title('Value counts of CUT')
sns.countplot(data = copydata, x = 'cut')
for container in ax[0].containers:
    ax[0].bar_label(container, color = 'black', size = 15)
    
plt.subplot(132)
plt.title('Value counts of COLOR')
sns.countplot(data = copydata , x = 'color')
for container in ax[1].containers:
    ax[1].bar_label(container, color = 'black', size = 15)

plt.subplot(133)
plt.title('Value counts of CLARITY')
sns.countplot(data = copydata, x = 'clarity')
for container in ax[2].containers:
    ax[2].bar_label(container, color = 'black', size = 15)
    
plt.show()
```
